#### Let's First handle the missing data.

Reading about NDVI, We understand that it is seasonal, So I'll be replacing the NA values in each row with their seasonal mean 

Considering 
- Summer => June, July and August
- Winter => December, January and February
- Autumn => September, October and December
- Spring => March, April and May 

```python
df_winter = df[['Unnamed: 0', 'ID', 'class']].copy()
df_spring = df[['Unnamed: 0', 'ID', 'class']].copy()
df_summer = df[['Unnamed: 0', 'ID', 'class']].copy()
df_autumn = df[['Unnamed: 0', 'ID', 'class']].copy()
```

```python
for col in df.columns:
    month = col[4:6] 
    
    if month in ["12", "01", "02"]:
        df_winter = pd.concat([df_winter, df[col]], axis=1)
        print("Added to winter dataframe", col[:8])
    elif month in ["03", "04", "05"]:
        df_spring = pd.concat([df_spring, df[col]], axis=1)
        print("Added to spring dataframe", col[:8])
    elif month in ["06", "07", "08"]: 
        df_summer = pd.concat([df_summer, df[col]], axis=1)
        print("Added to summer dataframe", col[:8])
    elif month in ["09", "10", "11"]:
        df_autumn = pd.concat([df_autumn, df[col]], axis=1)
        print("Added to autumn dataframe", col[:8])
    else:
        pass
```

```python
df_winter.info()
```

```python
df_summer
```

```python
# List of columns to exclude
exclude_cols = ['Unnamed: 0', 'ID', 'class']


mean_cols_winter = df_winter.columns.difference(exclude_cols)
mean_cols_summer = df_summer.columns.difference(exclude_cols)
mean_cols_autumn = df_autumn.columns.difference(exclude_cols)
mean_cols_spring = df_spring.columns.difference(exclude_cols)


row_means_winter = df_winter[mean_cols_winter].mean(axis=1)
row_means_summer = df_summer[mean_cols_summer].mean(axis=1)
row_means_autumn = df_autumn[mean_cols_autumn].mean(axis=1)
row_means_spring = df_spring[mean_cols_spring].mean(axis=1)
```

```python
row_means_summer
```

```python
# Replace NaNs with row-wise mean in the selected columns
df_winter[mean_cols_winter] = df_winter[mean_cols_winter].apply(
    lambda row: row.fillna(row.mean()), axis=1
)
df_summer[mean_cols_summer] = df_summer[mean_cols_summer].apply(
    lambda row: row.fillna(row.mean()), axis=1
)
df_autumn[mean_cols_autumn] = df_autumn[mean_cols_autumn].apply(
    lambda row: row.fillna(row.mean()), axis=1
)
df_spring[mean_cols_spring] = df_spring[mean_cols_spring].apply(
    lambda row: row.fillna(row.mean()), axis=1
)
```

```python

shared_cols = ['Unnamed: 0', 'ID', 'class']

# Get non-shared columns from each dataframe
winter_data = df_winter.drop(columns=shared_cols)
spring_data = df_spring.drop(columns=shared_cols)
summer_data = df_summer.drop(columns=shared_cols)
autumn_data = df_autumn.drop(columns=shared_cols)


df_final = pd.concat([df_winter[shared_cols], winter_data, spring_data, summer_data, autumn_data], axis=1)
```
